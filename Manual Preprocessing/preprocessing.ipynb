Step 4 - Replace the missing values

Categorical (Text) - Mode
Continuous (numerical) - Mean, Median

```python
df['AirBags'].mode()
```

```python
df['AirBags'].value_counts()
```

Function called as df.fillna(value)

```python
def replacer(df):
    cat = list(df.columns[df.dtypes=='object'])
    for i in df.columns:
        if i in cat:
            md = df[i].mode()[0]
            df[i] = df[i].fillna(md)
        else:
            mn = df[i].mean()
            df[i] = df[i].fillna(mn)
    print('Missing values replaced')
replacer(df)
```

```python
df.to_csv("Cars93clean.csv", index=False)
```
